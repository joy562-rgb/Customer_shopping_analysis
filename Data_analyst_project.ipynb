```python
import pandas as pd

```

```python

df = pd.read_csv("customer_shopping_behavior (1).csv")

```

```python
df.describe(include="all")
```

```python
df.isnull().sum()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.mean()))
```

```python
df.isnull().sum()
```

```python
#snake casing jate kono space r caital letter na thake
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# create a column age_group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age', 'age_group']].head(10)
```

```python
# create column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[[ 'frequency_of_purchases','purchase_frequency_days']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python


df = df.drop('promo_code_used', axis=1)
```
